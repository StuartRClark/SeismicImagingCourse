```python
import numpy as np # import an numpy that allows us to work easily with arrays
import matplotlib.pyplot as plt # Import our plotting function

seismic_trace = np.array([ 0.0 ,  0.5,  0.0 ,  2.5, -0.5,  0.0 , -2.5,  0.6,  0.0 ,  3.0 ]) # our data
times = np.linspace(0.0,3.6,num=10)

# Create the new time array recorded at 0.01 ms
times_x40 = np.linspace(0,3.6,num=400) # note 400 points not 10

# Interpolate the data
new_seis = np.interp(times_x40,times,seismic_trace)

# Plot 1D seismic trace as a Variable Density Display
fig, ax = plt.subplots(1,2)
fig.suptitle("Variable Density Plot (2 colour scales)")

# Transpose the array to plot it vertically - also need to make it 2d for image show (imshow)
new_seis_t = np.transpose(np.atleast_2d(new_seis))

# Left hand image, grey scale
im1 = ax[0].imshow(new_seis_t, cmap = 'gray_r', extent = [0,0.5,3.6,0]) # Plot image
ax[0].set(xlabel=None,xticks=[],ylabel="Two-Way Time (ms)",title="grayscale") #Set axis
fig.colorbar(im1,ax=ax[0], location='bottom',label="Seismic amplitude") # Add colorbar

# Right hand image, seismic colours
im2 = ax[1].imshow(new_seis_t, cmap = 'seismic', extent = [0,0.5,3.6,0])
ax[0].set(xlabel=None,xticks=[],ylabel="Two-Way Time (ms)",title="seismic")
fig.colorbar(im2,ax=ax[1], location='bottom',label="Seismic amplitude")

label_x = 1.5
# Labelling the horizons
ax[0].axhline(y = 1.1, linestyle='-', linewidth='2.5', color='red') #horizontal line at y = 1.2
ax[1].axhline(y = 1.1, linestyle='-', linewidth='2.5', color='black') #horizontal line at y = 1.2
ax[0].text   (s = 'Horizon 1', x = label_x, y = 1.1, c = 'black')        #text at y = 11 so it's above the line and x = -2.5 so it's on the left side of the figure.

ax[0].axhline(y = 2.5, linestyle='-', linewidth='2.5', color='red')
ax[1].axhline(y = 2.5, linestyle='-', linewidth='2.5', color='black')
ax[0].text   (s = 'Horizon 2', x = label_x, y = 2.3, c = 'black')

ax[0].axhline(y = 3.5, linestyle='-', linewidth='2.5', color='red')
ax[1].axhline(y = 3.5, linestyle='-', linewidth='2.5', color='black')
ax[0].text   (s = 'Horizon 3', x = label_x, y = 3.5, c = 'black');

#Display the plot
plt.show()
```
