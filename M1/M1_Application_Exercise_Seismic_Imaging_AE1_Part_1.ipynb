In the image below, 3 trucks are sending and receiving seismic signals. Signals bounce from interfaces between layers somewhere below the truck. Although the signal appears to bounce from immediately below the truck, it can bounce from anywhere at the same travel-time and bounce back to the truck, as long as the interface is at 90 degrees to the ray. Hence we don't know the actual angle of the layers based on this simple example (we usually have many more sensors so that we can tell). In this example, we assume that the layers are flat, but they can be of any shape as long as somewhere they are 90 degrees to the path from the truck.
