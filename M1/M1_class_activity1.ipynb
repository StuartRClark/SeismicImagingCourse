```python
import numpy as np # import an numpy that allows us to work easily with arrays
import matplotlib.pyplot as plt # Import our plotting function


seismic_trace = np.array([ 0.0 ,  0.5,  0.0 ,  2.5, -0.5,  0.0 , -2.5,  0.6,  0.0 ,  3.0 ]) # our data

# Plot seismic trace

# Setup an array to represent the timings of data point on our seismic array
times = np.linspace(0.0,3.6,num=10)

# Create a blank figure and axes
fig, ax = plt.subplots()

# Plotting the data
ax.plot(seismic_trace,times)

# Setup axes labels
ax.set(xlabel="Amplitude",ylabel="Two-Way Time (ms)",title="Seismic Wiggle Plot")
ax.invert_yaxis()  # plot from top to bottom because the deepest reflections take the longest to arrive back

# Labelling the horizons
plt.axhline(y = 1.2, linestyle='--', linewidth='0.5', color='grey')#horizontal line at y = 12
plt.text   (s = 'Horizon 1', x = -2.5, y = 1.1, c = 'grey')        #text at y = 11 so it's above the line and x = -2.5 so it's on the left side of the figure.
plt.axhline(y = 2.4, linestyle='--', linewidth='0.5', color='grey')
plt.text   (s = 'Horizon 2', x = 2, y = 2.3, c = 'grey')
plt.axhline(y = 3.6, linestyle='--', linewidth='0.5', color='grey')
plt.text   (s = 'Horizon 3', x = -2.5, y = 3.5, c = 'grey');
```
